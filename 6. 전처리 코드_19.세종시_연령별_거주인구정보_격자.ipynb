```python
import geopandas as gpd
import pandas as pd
import numpy as np
from geoband.API import *
GetCompasData('SBJ_2102_001', '19', '19.세종시_연령별_거주인구정보_격자.geojson')
```

```python
people = gpd.read_file(r'19.세종시_연령별_거주인구정보_격자.geojson')
```

```python
print(people.shape)
print(people.columns)
people.head()
```

```python
# 결측값으로 이루어진 행을 제거합니다.
people = people[people.isnull().sum(axis=1)!=36]
```

```python
# pandas DataFrame으로 변환합니다.
people = pd.DataFrame(people)
```

```python
# 각 연령별로, 연도별 거주인구수의 평균을 구해서 열로 추가합니다.

for i in range(20,110,10):
    if i != 100:
        people['mean_{}'.format(i)] = np.nanmean(people[['201710_{}대_거주인구수'.format(i),
                                                         '201810_{}대_거주인구수'.format(i),
                                                         '201910_{}대_거주인구수'.format(i),
                                                         '202010_{}대_거주인구수'.format(i)]].values,
                                                 axis=1,keepdims=True)
    else:
        people['mean_{}'.format(i)] = np.nanmean(people[['201710_{}세이상_거주인구수'.format(i),
                                                         '201810_{}세이상_거주인구수'.format(i),
                                                         '201910_{}세이상_거주인구수'.format(i),
                                                         '202010_{}세이상_거주인구수'.format(i)]].values,
                                                 axis=1,keepdims=True)
```
